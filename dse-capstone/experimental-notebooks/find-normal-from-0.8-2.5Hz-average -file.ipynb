```python
import pandas as pd
```

```python
df = pd.read_csv("../datasets/fft-min-max-on-average-0.8-2.5hz.csv") 
df.head()
```

```python
MIN_HR, MAX_HR = 0.048, 0.150

status = []
ncount = 0; 
for i, row in df.iterrows():
    if row['Min fft'] > MIN_HR  and row['Max fft'] < MAX_HR:
        status.append('Normal')
        ncount = ncount+1
    else:
        status.append('Not Normal')
df['Status'] = status

print(ncount)
```
