```python
import scipy.io as sio
from scipy.fft import fft
import os,glob
import math
import numpy as np
import pandas as pd
import os,glob
import pyexcel


def get_neck_files(dirname, filepatern):
    files = []
    for filename in glob.iglob(os.path.join(dirname, filepatern), recursive=True):
        files.append(filename)
    return files

folder_paths = ['../datasets/Age_A/Age_A_Male/', '../datasets/Age_A/Age_A_Female/',
               '../datasets/Age_B/Age_B_Male/', '../datasets/Age_B/Age_B_Female/',
               '../datasets/Age_C/Age_C_Male/', '../datasets/Age_C/Age_C_Female/']
age_group = []
sex = []
sub = []
side = []
filename = []
pats = ['**/*_R_*.mat', '**/*_L_*.mat']

Hz_Lthr, Hz_Hthr = 0, 10
for folder_path in folder_paths:
    for dirname in glob.iglob(os.path.join(folder_path, '**/')):
        for pat in pats:
            files = get_neck_files(dirname, pat)
            for file in files:
                parts=file.split('/')
                age_group.append(parts[2])
                sex.append(parts[3][6:])
                sub.append(parts[4])
                side.append('Neck'+pat[4:6])
                filename.append(file)
                
data = {
        'Age Group': age_group,
        'Sex': sex,
        'Side': side,
        'Sub': sub,
        'File': filename
       }  
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
df.to_csv('../datasets/laser_ds.csv',index = False, header=True)
```
