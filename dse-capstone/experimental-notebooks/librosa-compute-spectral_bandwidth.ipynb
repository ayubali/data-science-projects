```python
from mat4py import loadmat
import librosa
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../datasets/laser_ds.csv')
df.head()
```

```python
df.shape
```

```python
def get_signal(file):
    signal =  loadmat(file)
    data = [item for sublist in signal['data'] for item in sublist]
    y=np.array(data)
    sr=signal['Fs']
    return y, sr

def avg(lst): 
    return sum(lst) / len(lst)

spectral_bandwidth_avgs = []

for i, row in df.iterrows():
    y, sr = get_signal(row['File'])
    spectral_bandwidth_values = librosa.feature.spectral_bandwidth(y=y, sr=sr)    
    spectral_bandwidth_values= [item for sublist in spectral_bandwidth_values for item in sublist]
    spectral_bandwidth_avg = avg(spectral_bandwidth_values)
    spectral_bandwidth_avgs.append(spectral_bandwidth_avg)
    print('Processing is done for {} {}'.format(i,row['File']))
```

```python
df['spectral_bandwidth_avg'] =  spectral_bandwidth_avgs
```

```python
df.head()
```

```python
grouped_df = df.groupby(by=["Age Group","Sex","Sub"])
```

```python
rows = []
for name, group in grouped_df: 
    n_row = list(name)
    gr_avg = 0
    gr_count = 0
    for i, row in group.iterrows():
        gr_avg += row['spectral_bandwidth_avg']
        gr_count = gr_count + 1

    n_row.append(round(gr_avg/gr_count, 4))
    rows.append(n_row)
    

df_spectral_bandwidth = pd.DataFrame(rows);
df_spectral_bandwidth.head()
```

```python
df_spectral_bandwidth.columns = ["Age Group","Sex","Sub", "Spectral bandwidth avg"]
df_spectral_bandwidth.head()
```

```python
df_spectral_bandwidth.to_csv('../datasets/ds_spectral_bandwidth.csv',index = False, header=True)
```
