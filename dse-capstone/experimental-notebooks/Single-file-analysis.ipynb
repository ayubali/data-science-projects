```python
import scipy.io as sio
from scipy.fft import fft
import os,glob
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mat4py import loadmat
```

```python
import scipy.io as sio

def wfft_hz(filename, Hz_Hthr):
    data = loadmat(filename)
    fft_data = fft(data['data'])
    NFFT = len(data['data'])
    fft_data = abs(fft_data[0:NFFT])
    f = data['Fs']/2*np.linspace(0, 1, NFFT)
    idx = [i for i, e in enumerate(f) if e <= Hz_Hthr]
    X=f[idx]
    y=fft_data[idx]
    return X, y
```

```python
file = '../datasets/Age_A/Age_A_Female/Sub 25/25Neck_L_A_1.mat'
[X,y] =  wfft_hz('../datasets/Age_A/Age_A_Female/Sub 25/25Neck_L_A_1.mat', 10)
plt.figure(figsize=(10, 5), dpi=80)
plt.plot(X, y)
plt.ylabel('FFT magnitude')
plt.xlabel('Hz')
#plt.text(2000000, 0.9, "Min:"+str(min(y))+"  Max:"+str(max(y))  ,fontsize=16, bbox=dict(facecolor='red', alpha=0.5))
plt.show()
```
