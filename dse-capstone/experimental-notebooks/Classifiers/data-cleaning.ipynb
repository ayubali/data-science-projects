```python
import pandas as pd
```

```python
df = pd.read_csv("../../datasets/ds_hr_with_hz_derived.csv")
df.head()
```

```python
df1 = pd.read_csv("../../datasets/ds_hr_with_hz.csv")
df1.head()
```

```python
df1 = df1.join(df[['Health issue', 'Pulse', 'Level']])
df1.head()
```

```python
df1.to_csv('../../datasets/ds-with-all-features.csv', index=False)
```

```python
df1.columns
```

```python
df1 = df1.drop(['Age Group', 'Sex', 'Sub', 'Health issue', 'Pulse', 'Heart Beat 0.8~2.5hz'],axis=1)
```

```python
df1.head()
```

```python
df1.isna().sum()
```

```python
df1.dropna(how='any', subset=['Level'], inplace=True)
```

```python
df1= df1.reset_index(drop=True)
df1
```

```python
df1['Level'] = df1['Level'].apply(lambda x: 1 if x == 'abnormal' else 0)
df1.head()
```

```python
target_count = df1['Level'].value_counts()
print('Normal:', target_count[0])
print('Not Normal:', target_count[1])
```

```python
target_count
```

```python
df1= df1.rename(columns={"Level": "Not Normal"})
df1.head()
```

```python
df1.to_csv('../../datasets/clean-ds.csv', index=False)
```
