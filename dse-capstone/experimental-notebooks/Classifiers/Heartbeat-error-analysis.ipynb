```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../../datasets/ds-with-all-features.csv")
df.head()
```

```python
df.dropna(how='any', subset=['Level'], inplace=True)
```

```python
df=df[['Heart Beat 0.8~2.5hz', 'Heart Beat 0.8~2.5hz from Avg', 'Pulse']]
```

```python
df.head()
```

```python
errors_average=[]
errors_clustering=[]
for i, row in df.iterrows():
    avg_error = (abs(row['Heart Beat 0.8~2.5hz from Avg']-row['Pulse'])/row['Pulse'])*100
    clustering_error = (abs(row['Heart Beat 0.8~2.5hz']-row['Pulse'])/row['Pulse'])*100
    errors_average.append(round(avg_error,2))
    errors_clustering.append(round(clustering_error,2))
    
df['Error from Avg Method']= errors_average
df['Error from clustering Method']= errors_clustering
```

```python
df
```

```python
df_sum=df[['Error from Avg Method','Error from clustering Method']].sum()
```

```python
df_avg=(df_sum/df.shape[0])
df_avg
```

```python
print('Average error from Average method {}%'.format(round(df_avg[0],2)))
print('Average error from Clustering method {}%'.format(round(df_avg[1],2)))
```
