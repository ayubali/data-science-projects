```python
import scipy.io as sio
from scipy.fft import fft
import os,glob
import math
import numpy as np
import pandas as pd
import os,glob
import pyexcel
import signal_utll
```

```python
df = pd.read_csv('../datasets/laser_ds.csv')
df.head()
```

```python
top_peaks_array=[]
secs=60 # 1 mins
Hz_Lthr, Hz_Hthr, no_peaks = 0, 0.7, 5

for i, row in df.iterrows():
    hz_peaks = signal_utll.find_top_peaks_hz(row['File'], Hz_Lthr, Hz_Hthr, no_peaks)
    top_peaks_array.append(hz_peaks)

df['Peaks_hz'] = top_peaks_array
```

```python
df.head()
```

```python
grouped_df = df.groupby(by=["Age Group","Sex","Sub"])
```

```python
rows = []
for name, group in grouped_df: 
    n_row = list(name)
    gr_peaks = []
    for i, row in group.iterrows():
            gr_peaks.append(row['Peaks_hz'])
    gr_peaks_list = [item for sublist in gr_peaks for item in sublist]
    max_peak_cluster_center = signal_utll.do_clusters_and_find_max_center(gr_peaks_list, 5)
    n_row.append(round(max_peak_cluster_center, 2))
    rows.append(n_row)
    

df_Peaks= pd.DataFrame(rows);
df_Peaks
```

```python
df_Peaks.columns = ["Age Group","Sex","Sub", "Peak Hz in 0~0.7hz"]
df_Peaks.head()
```

```python
df_Peaks.to_csv('../datasets/ds_0_07Hz.csv',index = False, header=True)
```
